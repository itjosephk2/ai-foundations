# Sigmoid activation function (NNFS — Chapter 3)

This notebook implements the **Sigmoid activation function** and plots it using `matplotlib`.

The Sigmoid (logistic) function is defined as:

$sigma(x) = \frac{1}{1 + e^{-x}}$

It:

- Squashes any real-valued input to the range **(0, 1)**  
- Passes through **0.5** when \(x = 0\)  
- Becomes very close to 0 for large negative inputs and very close to 1 for large positive inputs  

Historically it was widely used as a hidden-layer activation, but nowadays it's more common in
output layers (e.g. binary classification).

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    """Sigmoid (logistic) activation function.
    
    Accepts scalars or NumPy arrays and returns values in (0, 1).
    """
    x = np.array(x)
    return 1 / (1 + np.exp(-x))

# Quick sanity check
test_values = np.array([-10.0, -2.0, -1.0, 0.0, 1.0, 2.0, 10.0])
print("x:         ", test_values)
print("sigmoid(x):", np.round(sigmoid(test_values), 4))
```

```python
# Plot the Sigmoid activation function

x = np.linspace(-6, 6, 400)
y = sigmoid(x)

plt.figure(figsize=(6, 4))
plt.plot(x, y, c='#cc5500')
plt.xlabel("x")
plt.ylabel("σ(x)")
plt.title("Sigmoid activation function")

# Draw axes like in the book-style diagrams
# lighter, semi-transparent axes
plt.axhline(0, color="tab:blue", alpha=0.5, linewidth=1)
plt.axvline(0, color="tab:blue", alpha=0.5, linewidth=1)

plt.ylim(-4, 4)
plt.grid(True)
plt.show()
```
