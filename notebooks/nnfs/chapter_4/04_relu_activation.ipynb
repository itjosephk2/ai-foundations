The **Rectified Linear Unit (ReLU)** is defined as:

$
y = 
\begin{cases}
x & x > 0 \\
0 & x \le 0
\end{cases}$

So it's just the line \(y = x\) for positive inputs, and **clipped to 0** for negative inputs.

ReLU is extremely popular because:

- it's very cheap to compute (just a max between 0 and \(x\)),  
- it keeps things almost linear while still introducing nonlinearity at 0,  
- it avoids the strong squashing of sigmoid in the positive range (less saturation, better gradients).

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    """ReLU activation function.
    
    For each element x:
    - returns 0 if x <= 0
    - returns x if x > 0
    Accepts scalars or NumPy arrays.
    """
    x = np.array(x)
    return np.maximum(0, x)

# Quick sanity check
test_values = np.array([-5.0, -2.0, -0.5, 0.0, 0.5, 2.0, 5.0])
print("x:       ", test_values)
print("ReLU(x):", relu(test_values))
```

```python
# Plot the ReLU activation function

x = np.linspace(-6, 6, 400)
y = relu(x)

plt.figure(figsize=(6, 4))

# Main ReLU line, darker orange, like the book style
plt.plot(x, y, c="#cc5500")

# Semi-transparent axes
plt.axhline(0, color="tab:blue", alpha=0.7, linewidth=1)
plt.axvline(0, color="tab:blue", alpha=0.7, linewidth=1)

plt.xlabel("x")
plt.ylabel("y")
plt.title("ReLU activation function")

plt.grid(True)
plt.ylim(-2, 10)  # enough room to see the 'kink' at 0
plt.show()
```
