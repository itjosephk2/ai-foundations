This notebook implements the **linear activation function** and plots it using `matplotlib`.

For the linear activation, the output equals the input:

\[
y = x
\]

This is just the equation of a straight line through the origin.

In neural networks, a linear activation is often used in the **output layer of regression models**, 
where we want the network to predict a real-valued quantity (e.g. a price, a temperature, etc.).


```python
import numpy as np
import matplotlib.pyplot as plt

def linear(x):
    """Linear activation function (identity).
    
    Simply returns the input.
    Accepts scalars or NumPy arrays.
    """
    return np.array(x)

# Quick sanity check
test_values = np.array([-5.0, -2.0, -0.5, 0.0, 0.5, 2.0, 5.0])
print("x:        ", test_values)
print("linear(x):", linear(test_values))
```

```python
# Plot the linear activation function y = x

x = np.linspace(-6, 6, 400)
y = linear(x)

plt.figure(figsize=(6, 4))
plt.plot(x, y, c='#cc5500')  # y = x
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear activation function (y = x)")
plt.ylim(-6, 6)

# Draw x- and y-axes to resemble the book's style
plt.axhline(0)
plt.axvline(0)

plt.grid(True)
plt.show()
```
