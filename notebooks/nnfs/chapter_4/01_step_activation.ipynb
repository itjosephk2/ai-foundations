# Step activation function (NNFS â€” Chapter 3)

This notebook implements the **step activation function** and plots it using `matplotlib`.

Definition (as in the book):

- If the neuron's pre-activation value (\(x\)) is greater than 0, the neuron **fires**: \(y = 1\)
- If \(x \le 0\), the neuron does **not** fire: \(y = 0\)

\[
y = 
\begin{cases}
1 & x > 0 \\
0 & x \le 0
\end{cases}
\]


```python
import numpy as np
import matplotlib.pyplot as plt

def step(x):
    """Step activation function.
    
    Returns 1 where x > 0, else 0.
    Accepts scalars or NumPy arrays.
    """
    x = np.array(x)
    return np.where(x > 0, 1, 0)

# Quick sanity check
test_values = np.array([-2.0, -0.5, 0.0, 0.5, 3.0])
print("x:      ", test_values)
print("step(x):", step(test_values))
```

```python
# Plot the step function

x = np.linspace(-5, 5, 400)
y = step(x)

plt.figure(figsize=(6, 4))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Step activation function")
plt.ylim(-0.5, 1.5)
plt.grid(True)
plt.show()
```
