Indexing and slicing

```python
import numpy as np
```

```python
twodim_arr=np.reshape(np.arange(12),(3,4))
twodim_arr
```

```python
twodim_arr[1,1]
twodim_arr[1]
```

```python
threedim_arr=np.reshape(np.arange(3*4*5),(3,4,5))
threedim_arr
```

```python
threedim_arr[0,2,3]
```

```python
threedim_arr[2,-1,1]
```

```python
onedim_arr=np.arange(10)
onedim_arr
```

```python
onedim_arr[2:6]
```

```python
onedim_arr[:5]
```

```python
onedim_arr[-3:]
```

```python
onedim_arr[::3]
```

```python
twodim_arr[1:,1:]
```

```python
twodim_arr[1,:]
```

```python
twodim_arr[:,2]
```
