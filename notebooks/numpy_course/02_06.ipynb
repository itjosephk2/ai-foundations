Finding he shape and size of an array

```python
import numpy as np
```

```python
first_arr=np.arange(20)
first_arr
```

```python
second_arr=np.linspace((1,2),(10,20),10)
second_arr
```

```python
third_arr=np.full((2,2,2),10)
third_arr
```

```python
np.shape(first_arr)
```

```python
np.shape(second_arr)
```

```python
np.shape(third_arr)
```

```python
np.size(first_arr)
```

```python
np.size(second_arr)
```

```python
np.size(third_arr)
```
