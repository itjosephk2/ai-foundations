Reversing an array

```python
import numpy as np
```

```python
arr_one_dim=[10,1,9,2,8,3,7,4,6,5]
arr_one_dim
```

```python
arr_one_dim[::-1]
```

```python
np.flip(arr_one_dim)
```

```python
arr_two_dim=np.arange(9).reshape(3,3)
arr_two_dim
```

```python
np.flip(arr_two_dim, axis=1)
```

```python
np.flip(arr_two_dim)
```

```python
arr_three_dim=np.arange(24).reshape(2,3,4)
arr_three_dim
```

```python
np.flip(arr_three_dim, 1)
```

```python
np.flip(arr_three_dim, 2)
```
