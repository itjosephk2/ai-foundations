Transpose like operations

```python
import numpy as np
```

```python
first_2d_arr=np.arange(12).reshape((3,4))
first_2d_arr
```

```python
np.transpose(first_2d_arr)
```

```python
second_2d_arr=np.arange(6).reshape(3,2)
second_2d_arr
```

```python
np.transpose(second_2d_arr, (1,0))
```

```python
first_3d_arr=np.arange(24).reshape(2,3,4)
first_3d_arr
```

```python
np.moveaxis(first_3d_arr,0,-1)
```

```python
np.swapaxes(first_3d_arr,0,2)
```
