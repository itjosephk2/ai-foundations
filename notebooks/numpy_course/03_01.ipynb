Adding, removing and sorting elements

```python
import numpy as np
```

```python
first_arr=np.array([1,2,3,5])
print('Array')
print(first_arr)
new_first_arr=np.insert(first_arr,3,4)
print('Insert')
print(new_first_arr)
```

```python
second_arr=np.array([1,2,3,4])
print('Array')
print(second_arr)
new_second_arr=np.append(second_arr,5)
print('Element appended')
print(new_second_arr)
```

```python
third_arr=np.array([1,2,3,4,5,6])
print('Array')
print(third_arr)
del_arr=np.delete(third_arr,5)
print('Element 5 deleted')
print(del_arr)
```

```python
integers_arr=np.random.randint(0,20,20)
print('Unsorted array')
print(integers_arr)
print('Sorted Array')
print(np.sort(integers_arr))
```

```python
integers_2dim_arr=np.array([[3,2,5,7,4],[5,0,8,3,1]])
print('Unsorted 2D Array')
print(integers_2dim_arr)
print('Sorted 2D Array')
print(np.sort(integers_2dim_arr))
```
