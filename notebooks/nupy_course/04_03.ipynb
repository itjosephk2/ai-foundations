Aggregate functions

```python
import numpy as np
```

```python
first_arr=np.arange(10,110,10)
first_arr
```

```python
second_arr=np.arange(10,100,10).reshape(3,3)
second_arr
```

```python
third_arr=np.arange(10,110,10).reshape(2,5)
third_arr
```

```python
first_arr.sum()
```

```python
second_arr.sum()
```

```python
third_arr.sum()
```

```python
second_arr.sum(axis=0)
```

```python
second_arr.sum(axis=1)
```

```python
first_arr.prod()
```

```python
second_arr.prod()
```

```python
third_arr.prod()
```

```python
third_arr.prod(axis=0)
```

```python
np.average(first_arr)
```

```python
np.average(second_arr)
```

```python
np.average(third_arr)
```

```python
np.min(first_arr)
```

```python
np.max(first_arr)
```

```python
np.mean(first_arr)
```

```python
np.std(first_arr)
```
