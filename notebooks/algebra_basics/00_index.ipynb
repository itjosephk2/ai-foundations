# Algebra Foundations — Index  

This folder is my **Basic Algebra** sandbox.  
Goal: get fluent with numbers, expressions, and simple equations so later stuff (calculus, linear algebra, ML) feels like a natural extension, not a brick wall.

---

## How to use these notebooks  

1. Start at the lowest number you’re not comfortable with.
2. Do problems by hand first.
3. Use Python cells to check answers and experiment.
4. Add a short note at the end of each notebook: *what felt easy / what felt sticky*.
