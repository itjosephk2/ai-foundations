# algebra foundations — index  

this folder is my **basic algebra** track.

goal: get properly fluent with core algebra so that later stuff  
(linear algebra, calculus, probability, ml) feels like a continuation,  
not a restart.

---

## how this folder works  

each notebook is a small “chapter”:

- read the short intro
- do problems by hand first (paper / tablet / whatever)
- use python to check answers and experiment
- leave a few notes at the end for future me

i should be able to come back in 6–12 months, skim this folder,  
and get my algebra brain back online quickly.
