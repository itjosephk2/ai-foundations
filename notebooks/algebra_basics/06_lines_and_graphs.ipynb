## 1. Points and the coordinate plane

A point is written as an ordered pair \((x, y)\):

- The first number is the **x-coordinate** (left-right).
- The second number is the **y-coordinate** (up-down).

Examples:

- (2, 1) → 2 units right, 1 unit up.
- (-3, 4) → 3 units left, 4 units up.
- (0, -2) → on the y-axis, 2 units down.

Even without drawing every time, I can still **think** in terms of where a point lies.

Quick mental checks:

- If \(x = 0\), the point lies on the **y-axis**.
- If \(y = 0\), the point lies on the **x-axis**.


## 2. The equation of a line: \(y = mx + b\)

A very common way to write a straight line is:

\[ y = mx + b \]

where:

- \(m\) is the **slope** (steepness and direction),
- \(b\) is the **y-intercept** (where the line crosses the y-axis, i.e. the value of y when x = 0).

Examples:

- \(y = 2x + 1\): slope 2, y-intercept 1.  
- \(y = -\frac{1}{2}x + 3\): slope -1/2, y-intercept 3.

Interpretation of slope \(m\):

- Positive \(m\): line goes **up** as x increases.
- Negative \(m\): line goes **down** as x increases.
- \(m = 0\): horizontal line.


## 3. Building a small table of values

One simple way to sketch a line is:

1. Pick a few x-values.
2. Compute the corresponding y-values using \(y = mx + b\).
3. Plot the points and connect them.

Example: \(y = 2x + 1\)

Choose x = -1, 0, 1, 2.

- x = -1 → y = 2(-1) + 1 = -2 + 1 = -1 → point (-1, -1)  
- x = 0 → y = 2(0) + 1 = 1 → point (0, 1)  
- x = 1 → y = 2(1) + 1 = 3 → point (1, 3)  
- x = 2 → y = 2(2) + 1 = 5 → point (2, 5)

Two points are enough to define a line, but more points help check for mistakes.

I can write them like:

- x: …  
- y: …  
- points: (x, y), …


```python
def line_values(m, b, xs):
    """Given slope m, intercept b, and a list of x-values, return (x, y) pairs."""
    points = []
    for x in xs:
        y = m * x + b
        points.append((x, y))
    return points

# Quick example: y = 2x + 1 for x = -1, 0, 1, 2
xs_example = [-1, 0, 1, 2]
points_example = line_values(2, 1, xs_example)
points_example
```

## 4. Slope as “rise over run”

Given two points on a line, say \((x_1, y_1)\) and \((x_2, y_2)\),  
the slope \(m\) is:

\[
m = \frac{y_2 - y_1}{x_2 - x_1}
\]

This is **rise over run**:

- **rise** = change in y = \(y_2 - y_1\)  
- **run** = change in x = \(x_2 - x_1\)

Examples:

1. From (1, 2) to (3, 6):

   \[
   m = \frac{6 - 2}{3 - 1} = \frac{4}{2} = 2
   \]

2. From (0, 4) to (2, 1):

   \[
   m = \frac{1 - 4}{2 - 0} = \frac{-3}{2} = -1.5
   \]

---

### Your turn — slope between two points

Find the slope of the line through:

1. (2, 5) and (4, 9)  
2. (-1, 3) and (3, 3)  
3. (0, -2) and (2, 1)

For each, decide whether the line is going uphill, downhill, or flat.


```python
def slope(p1, p2):
    """Return the slope between two points p1=(x1, y1), p2=(x2, y2)."""
    x1, y1 = p1
    x2, y2 = p2
    if x2 == x1:
        raise ValueError("Vertical line: slope is undefined (infinite).")
    return (y2 - y1) / (x2 - x1)

# Quick slope checks
print("Slope from (1, 2) to (3, 6):", slope((1, 2), (3, 6)))
print("Slope from (0, 4) to (2, 1):", slope((0, 4), (2, 1)))
```

## 5. Plotting lines with Python (visual check)

I can use Python and matplotlib to **visualise** some lines.  
This isn’t required for understanding, but it helps build intuition.

We’ll make a small helper to plot \(y = mx + b\) over a chosen x-range.


```python
import matplotlib.pyplot as plt

def plot_line(m, b, x_min=-10, x_max=10, num_points=100):
    xs = [x_min + (x_max - x_min) * i / (num_points - 1) for i in range(num_points)]
    ys = [m * x + b for x in xs]

    plt.figure()
    plt.axhline(0)  # x-axis
    plt.axvline(0)  # y-axis
    plt.plot(xs, ys)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(f"y = {m}x + {b}")
    plt.grid(True)
    plt.show()

# Example: plot y = 2x + 1
plot_line(2, 1)
```

## 6. Matching equations and graphs (conceptually)

Without always plotting, I can often **guess the shape**:

- Large positive slope (e.g. \(m = 3\)): steep uphill.
- Small positive slope (e.g. \(m = 0.5\)): gentle uphill.
- Negative slope (e.g. \(m = -2\)): downhill as x increases.
- \(m = 0\): horizontal line at y = b.

Examples:

- \(y = 3x - 1\): passes through (0, -1), goes up quickly.  
- \(y = -\frac{1}{2}x + 4\): passes through (0, 4), goes down gently.  
- \(y = 0x + 2\) or just \(y = 2\): horizontal line at y = 2.

## 7. From two points to an equation

Sometimes I’m given two points and asked for the line equation.

General plan:

1. Find the slope \(m\) using the two points.
2. Use one point and the formula \(y = mx + b\) to solve for \(b\).
3. Write the final equation \(y = mx + b\).

---

### Example 7.1

Find the equation of the line through (1, 2) and (3, 6).

1. Slope:

\[
m = \frac{6 - 2}{3 - 1} = \frac{4}{2} = 2
\]

2. Use point (1, 2) in \(y = mx + b\):

\[
2 = 2(1) + b \Rightarrow 2 = 2 + b \Rightarrow b = 0
\]

3. Equation: \(y = 2x\).
