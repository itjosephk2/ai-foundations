## 1. Basic powers

A power has the form \(a^n\):

- \(a\) is the **base**,
- \(n\) is the **exponent** (or power).

If \(n\) is a positive integer,

\[
a^n = \underbrace{a \cdot a \cdot \dotsm \cdot a}_{n\text{ times}}
\]

Examples:

- \(2^3 = 2 \cdot 2 \cdot 2 = 8\)
- \(5^1 = 5\)
- \((-3)^2 = 9\)
- \((-3)^3 = -27\)

## 2. Laws of exponents (same base)

For a non-zero base \(a\) and integers \(m, n\):

1. **Product rule**  
   \[ a^m \cdot a^n = a^{m + n} \]

2. **Quotient rule**  
   \[ \frac{a^m}{a^n} = a^{m - n} \quad (a \ne 0) \]

3. **Power of a power**  
   \[ (a^m)^n = a^{mn} \]

4. **Power of a product**  
   \[ (ab)^n = a^n b^n \]

These are just shorthand for repeated multiplication.

## 3. Zero and negative exponents

To keep the laws consistent, we define:

- **Zero exponent** (for \(a \ne 0\)):  
  \[ a^0 = 1 \]

Why? Using the quotient rule:

\[
\frac{a^3}{a^3} = a^{3-3} = a^0
\]

But also \(a^3 / a^3 = 1\), so \(a^0 = 1\).

---

- **Negative exponents** (for \(a \ne 0\)):  
  \[ a^{-n} = \frac{1}{a^n} \]

Example:

- \(2^{-3} = 1 / 2^3 = 1/8\)
- \(10^{-2} = 1 / 10^2 = 1/100\)

## 4. Scientific notation (base 10)

Scientific notation is a way to write very large or very small numbers using powers of 10.

General form:

\[
a \times 10^n
\]

where:

- \(1 \le |a| < 10\),  
- \(n\) is an integer.

Examples:

- 3000 = \(3 \times 10^3\)  
- 0.004 = \(4 \times 10^{-3}\)  
- 12 000 = \(1.2 \times 10^4\)
